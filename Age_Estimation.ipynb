```python
from deepface import DeepFace
import gradio as gr
import requests
import json
```

```python
def Age_Estimation(picture):
  obj = DeepFace.analyze(img_path=picture, actions=["age"])
  return obj["age"]

with gr.Blocks() as demo:
    pic = gr.Image(label="Image")
    sen = gr.Textbox(label="Age Estimation: ")
    greet_btn = gr.Button("Estimate")
    greet_btn.click(fn=Age_Estimation, inputs=pic, outputs=sen)

demo.launch()
```
